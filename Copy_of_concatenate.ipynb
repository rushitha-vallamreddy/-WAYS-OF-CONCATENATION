```python
import pandas as pd
data1={'Name':['jai','princi','gaurav','xnuj'],
       'Age':[27,24,22,32],
       'Address':['nagpur','kanpur','allahabad','kannuj'],
       'Qualification':['Msc','Ma','Mca','Phd']}

data2={'Name':['rushi','harshi','deekshi','kushi'],
       'Age':[17,14,12,52],
       'Address':['nagpur','kanpur','allahabad','kannuj'],
       'Qualification':['Btech','Ba','BcOM','Bhonest']}

df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[4,5,6,7])
print(df,"\n\n",df1)
res1=pd.concat([df,df1])
res1

```

```python
import pandas as pd
data1={'Name':['jai','princi','gaurav','xnuj'],
       'Age':[27,24,22,32],
       'Address':['nagpur','kanpur','allahabad','kannuj'],
       'Qualification':['Msc','Ma','Mca','Phd']}

data2={'Name':['rushi','harshi','deekshi','kushi'],
       'Age':[17,14,12,52],
       'Address':['nagpur','kanpur','allahabad','kannuj'],
       'Qualification':['Btech','Ba','BcOM','Bhonest']}

df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
res1=pd.concat([df,df1])
res1
res2=pd.concat([df,df1],axis=1,join='inner')
res2


```

```python
import pandas as pd
data1={'Name':['jai','princi','gaurav','xnuj'],
       'Age':[27,24,22,32],
       'Address':['nagpur','kanpur','allahabad','kannuj'],
       'Qualification':['Msc','Ma','Mca','Phd']}

data2={'Name':['rushi','harshi','deekshi','kushi'],
       'Age':[17,14,12,52],
       'Address':['nagpur','kanpur','allahabad','kannuj'],
       'Qualification':['Btech','Ba','BcOM','Bhonest']}

df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
res=pd.concat([df,df1],ignore_index=True)
res

```

```python
import pandas as pd
data1={'Key':['k0','k1','k2','k3'],
       'Name':['jai','princi','gaurav','xnuj'],
       'Age':[27,24,22,32],}


data2={'Key':['k0','k1','k2','k3'],
       'Address':['Nagapur','Kanpur','allahabad','kannuaj'],
       'Qualification':['Btech','Ba','BcOM','Bhons']}

df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res=pd.merge(df,df1,on='Key')
res

```
